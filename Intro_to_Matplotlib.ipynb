```python
#Matplot workflow example
#1.import matplotlib 

%matplotlib inline 
import matplotlib.pyplot as plt

#2. prepare data 
x = [1,2,3,4]
y = [11,22,33,44]

#set up plot object oriented way
fig, ax = plt.subplots(figsize=(10,2)) #width and height 

#3 Plot the data 
ax.plot(x, y)

#4. Customize the plot 
ax.set(title="Simple Plot", 
       xlabel = "x-axis",
       ylabel='y-axis')
#5.Save and show 
#fig.savefig("path")
plt.show()
```

```python
#Figure from a dictionary

nut_butter = {
    "Almond Butter":10,
    "Peaut Butter": 8,
    "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter.keys(),nut_butter.values())
ax.set(title="Nut Butter Prices", xlabel="Butter", ylabel="Prices");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter.keys()),list(nut_butter.values()))
ax.set(title="Nut Butter Prices", xlabel="Butter", ylabel="Prices");
```

```python
#Histogram 
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
ax.set(title="Normal Dist", xlabel="X", ylabel="Y");
```

```python
#subplots option 1  #using tuple 

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.barh(list(nut_butter.keys()),list(nut_butter.values()))
ax4.hist(np.random.randn(1000));
```

```python
#subplots option 2 #using list
fig,ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
ax[0, 0].plot(x,x/2)
ax[0, 1].scatter(np.random.random(10),np.random.random(10))
ax[1, 0].barh(list(nut_butter.keys()),list(nut_butter.values()))
ax[1, 1].hist(np.random.randn(1000));
```

Using another dataset - heart disease

```python
df1=pd.read_csv('https://raw.githubusercontent.com/kmshravani/LinearRegression/main/heart-disease.csv')
df1.head()
```

```python
df1['age'].plot.hist(bins=20);
#if you observe, it is a uniform dist curve ~ bell curve
#outliers present 3 std away from mean usually
```

```python
#df1.plot.hist(subplots=True);
df1.plot.hist(figsize=(10,30), subplots=True);
#will take sometime to load and not useful
```

```python
#using object oriented way to use matplotlib directly with pandas
#pyplot vs matplotlib OO method
#when plotting something quickly , okay to use pyplot method ~ df['Odometer (KM)'].plot.hist();
# when plotting advanced, use OO method 
```

```python
over_50 = df1[df1['age']>50]
len(over_50)
display(over_50.head())
```

```python
over_50.plot(kind='scatter',
             x='age',
            y='chol',
             c='target')
```

```python
#OO method mixed with pyplot
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
            y='chol',
             c='target',
             ax=ax); 
ax.set_xlim([45, 85])
```

```python
#OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))
#plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target']
                     );
#Customize The plot
ax.set(title="Heart Disease and Cholesterol level",
       xlabel="Age",
       ylabel='Cholesterol')
#Add a legend
ax.legend(*scatter.legend_elements(),title="Targettt")
#*scatter.legend_elements() - will go to scatter(which has plot) and get the legend elements which gets target values from df1

#add horizontal line
ax.axhline(over_50['chol'].mean(),
           linestyle='--');
```

```python
#add subplots to above figure
#subplot of chol, age, thalach

fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)
#----------------------------------------------
#ax0
#add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])
#customize ax0
ax0.set(title="Heart Disease and Cholesterol level",
       ylabel='Cholesterol') #xlabel="Age",
                             
#Add a legend
ax0.legend(*scatter.legend_elements(),title="Targetax0")
#add horizontal line
ax0.axhline(y=over_50['chol'].mean(),
           linestyle='--');
#----------------------------------------------
#ax1
#add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'])
#customize ax0
ax1.set(title="Heart Disease and Max Heart Rate(thalach)level",
        xlabel="Age",
       ylabel='Max Heart Rate(thalach)') # xlabel="Age",
#Add a legend
ax1.legend(*scatter.legend_elements(),title="Targetax1")
#add horizontal line
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle='--');

#Have one title for both plots
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')
```
